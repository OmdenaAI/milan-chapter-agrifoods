```python
col_list = list(data.columns)
```

```python
for col in col_list:
    if data[col].dtypes=='object':
        data[col] = data[col].fillna(data[col].mode()[0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for col_name in col_list:
    if(data[col_name].dtypes=='int64' or data[col_name].dtypes=='float64'):
        sns.boxplot(data[col_name])
        plt.xlabel(col_name)
        plt.ylabel('count')
        plt.show()
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
Q1 = data['TIME'].quantile(0.25)
Q3 = data['TIME'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['TIME'] >= Q1 - 1.5*IQR) & (data['TIME'] <= Q3 + 1.5*IQR)]

Q1 = data['Select time'].quantile(0.25)
Q3 = data['Select time'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['Select time'] >= Q1 - 1.5*IQR) & (data['Select time'] <= Q3 + 1.5*IQR)]

Q1 = data['Value'].quantile(0.25)
Q3 = data['Value'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['Value'] >= Q1 - 1.5*IQR) & (data['Value'] <= Q3 + 1.5*IQR)]
```
